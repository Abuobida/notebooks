## Resampling

We will use some code from Evan's [geocomputing course](http://agilegeoscience.com/courses). 

```python
import helen
```

The file headers contain information about the extents and sample intervals of the DEM grids.

```python
before_headers = helen.get_headers('data/st-helens_before.txt')
after_headers = helen.get_headers('data/st-helens_after.txt')
print before_headers
print after_headers
```

```python
dx, dy = helen.pixel_inc(after_headers)
print dx, dy
```

```python
from scipy.interpolate import RectBivariateSpline
```

```python
def resample(data, headers, delta_new):
    
    x_min = min(headers['x_extent'])
    x_max = max(headers['x_extent'])
    y_min = min(headers['y_extent'])
    y_max = max(headers['y_extent'])
    
    x_new = np.arange(x_min, x_max, delta_new)
    y_new = np.arange(y_min, y_max, delta_new)

    dx = ( max(headers['x_extent']) - min(headers['x_extent']) ) / float ( headers['Dimensions'][0] )
    dy = ( max(headers['y_extent']) - min(headers['y_extent']) ) / float ( headers['Dimensions'][1] )
    
    x = np.arange(x_min, x_max, dx)
    y = np.arange(y_min, y_max, dy)
    xx, yy = np.meshgrid(x,y)
    
    interpolator = RectBivariateSpline(x, y, data, kx=3, ky=3, s=0)
    new_data = interpolator(x_new, y_new)
    
    return new_data
```

```python
after_re = resample(after, after_headers, 10)
```

```python
before_re = resample(before, before_headers, 10)
```

```python
after_re.shape, before_re.shape
```

Ugh, right, look at the headers — we have a little more data in <code>after</code>. This is getting very hacky... 

```python
after_re = after_re[1:-1, 2:-2]
```
