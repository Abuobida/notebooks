```python
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
from geopandas import GeoSeries, GeoDataFrame

%matplotlib inline
```

```python
# initial conditions
xmi = 575000   # leftmost corner of grid (m)
ymi= 4710000   # bottommost corner of grid (m)
SL = 600       # shot line interval (m)
RL = 600       # receiver line interval (m)
si = 100       # source point interval (m)
ri = 100       # receiver interval (m)
x = 3000       # width of survey (m)
y = 1800       # height of survey (m)
```

```python
# Number of receiver lines and source lines
rlines = int(y/RL) + 1
slines = int(x/SL) + 1

# Put recevier lines East-West, and shot lines North South
rperline = int(x/ri) + 2
sperline = int(y/si) + 2

# offset the receivers relative to the sources
shiftx = -si/2.
shifty = -ri/2.
```

```python
# Find x,y coordinates of recs and shots
rcvrx = [xmi + station*ri + shiftx for line in range(rlines) for station in range(rperline)]
rcvry = [ymi + line*RL - shifty for line in range(rlines) for station in range(rperline)]

srcx = [xmi + line*SL for line in range(slines) for station in range(sperline)]
srcy = [ymi + station*si for line in range(slines) for station in range(sperline)] 
```

```python
# create figure

def plot_geoms(xcoords, ycoords, color = 'none', size = 50, alpha = 0.5):
    a = plt.scatter(xcoords, ycoords, 
        c = color, 
        s = size,
        marker = 'o', 
        alpha = alpha,
        edgecolor = 'none'
        )
    return a
```

```python
fig = plt.figure(figsize=(15,15))
plot_geoms(rcvrx, rcvry, 'b')
b = plot_geoms(srcx, srcy, 'r')
plt.axis('equal')
plt.grid()
```

### Using Shapely and GeoPandas

```python
# zip into x,y coordinates
rcvrxy = zip(rcvrx, rcvry)
srcxy = zip(srcx, srcy)

# Create a list of Points
rcvrs = [Point(x, y) for x, y in rcvrxy]
srcs = [Point(x, y) for x, y in srcxy]

# Put GeoSeries into a GeoDataFrame
sources = GeoDataFrame({'geometry' : srcs})
receivers =  GeoDataFrame({'geometry' : rcvrs})

# Remove the last two comments if you want ...
# ...to Save GeoDataFrame as Shapefile
# sources.to_file('sources.shp')
# receivers.to_file('receivers.shp')
```
