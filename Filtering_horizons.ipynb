Reading data
============

Now we can read the data into a Python variable called <code>data</code>:

```python
with open('Penobscot_HorB.txt') as f:
    data = f.read()
```

```python
print data[:100]
```

The square bracket notation is called a 'slice'. The *data* object is a string. The slice gives us the first 100 characters of it. Slicing, and the closely related indexing action, are very useful — and you'll see them a lot in Python code.

As expected, we have inline, xline, and two-way time. There is no header. We only really need the time values, so let's use slicing and indexing some more

We can actually make our life a bit easier by re-reading the file into a list, one line at a time. This is easy:

```python
with open('Penobscot_HorB.txt') as f:
    data = f.readlines()

data[:5]
```

Those <code>\t</code> things are tabs. The <code>\n</code> characters are newlines. Even though they have those backslashes, they are single characters. One thing to note here — these are text objects (strings), not numbers. 

Let's take one element (the 4th one, index 3), and split it based on whitespace characters. 

```python
data[3].split()
```

We can go further: let's select the z-value by its index:

```python
data[3].split()[2]
```

Let's make a loop to take each line in turn, and split it up. We'll also cast the strings into numbers. The command int('5') will make an integer; if we need decimal precision, we'll make a float with float('3.14159'). If we didn't do this step, we'd end up with strings.

```python
# First make a list to hold the values
twtt = []

# Then iterate over the lines of the file (now in our list data)
for d in data:
    # Split each element, and append each third chunk to our list
    inline = d.split()[0]
    xline = d.split()[1]
    z_value = d.split()[2]
    
    # Convert them to numbers
    inline = int(inline)
    xline = int(xline)
    z_value = float(z_value)
    
    sample = [inline, xline, z_value]
    
    twtt.append(sample)

twtt[:5]
```

Notice that we only exported values where there is data — inlines 1003 to 1597, and xlines 1009 to 1471. (We know from OpendTect that the extents of the survey data are: inlines 1000 to 1600, and xlines 1000 to 1481.) Also notice that the first row in our dataset is near the lower-left corner of the survey. 

To keep things easy later on, we'll use the extents of the horizon from the OpendTect export dialog (above), and preserve this shape. 

```python
inlines = 1597 - 1003 + 1
xlines = 1471 - 1009 + 1

print inlines, 'inlines'
print xlines, 'crosslines'
```

So we'll want a data object — an array — that has 595 rows and 463 columns. We'll make that in a minute. 

Mapping our data into that array will be much easier if we convert those inlines and crosslines to 'coordinates', called indices, in the array. 

```python
for sample in twtt:
    sample[0] = 595 - (sample[0] - 1002)
    sample[1] = sample[1] - 1008
    
twtt[:5]
```

Now we have a Python *list* but we really want a NumPy array. NumPy is just a Python library (a bunch of commands, basically) that makes mathematics on big datasets faster. We could use Python's standard tools, but why make things more difficult for ourselves? To use a library like NumPy, we have to import it first.

```python
import numpy
```

First we'll make an empty array to hold our data:

```python
horizon = numpy.empty((inlines, xlines))
print horizon
print horizon.shape
```

Now we can cast the list we made into the array.

```python
for sample in twtt:
    inline = sample[0]
    xline = sample[1]
    z_value = sample[2]
    
    # We have to subtract 1 to allow for 0-based indexing
    horizon[inline - 1, xline - 1] = z_value

horizon
```

Notice 2 things when we look at the array after this operation:

- NumPy arrays look a lot like lists. 
- NumPy conveniently summarizes the array, so we don't get lines and lines of data. 
- Look back at the contents of the file and the initial twtt variable. Those might look like numbers, but they were actually text, or *strings*. The quote marks are the giveaway. When I ask for a datatype (*dtype*) of float32, NumPy conventiently casts the strings into 32-bit floating point numbers for me. 
- There is a problem: the array is one-dimensional, but it should be a 2D array representing a horizon. 

Let's deal with the shape of the array right away. I need to know the number of inlines and crosslines, which I can get from OpendTect (or I could go back to my *data* variable and extract the inlines and crosslines and find the min and max with Python).

```python
print horizon.shape
print inlines * xlines
```

So our current array has 275,485 elements.

What is convolution?
====================
Convolution is an algorithm: a stepwise process with some input and some transformed or generated output. Let's look at a 1D example. We'll convolve a simple pair of numbers, which you can think of as two spikes, with a special function that smoothes spikes out (a filter):

```python
import numpy
my_spikes = numpy.zeros(19)
my_spikes[5] = 1
my_spikes[13] = -2

my_filter = numpy.array([0.05, 0.2, 0.5, 0.2, 0.05])

x = numpy.arange(len(my_spikes)) # for the plot

plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(my_spikes)
plt.subplot(1,2,2)
plt.plot(my_filter)
plt.show()
```

Now we can do the convolution. It happens one step at a time. At each step, we multiply the spiky array by the filter, and sum the results. Notice that to evaluate the first element in the array, we'll either need to pad with zeros, or cut off the first two elements. We'll cut them off. 

```python
result = numpy.zeros_like(my_spikes[2:-2])  # Copy the 3 to the (n-2)th elements of the array
for i in numpy.arange(len(result)):
    for j in numpy.arange(len(my_filter)):
        result[i] += my_spikes[(i + 2) +  (j - 2)] * my_filter[j]

print result

plt.plot(my_spikes[2:-2])
plt.plot(result)
plt.show()
```

The result is essentially just the sum of the kernels, scaled by the spikes. 

Building a kernel
=================
Before we can do any work on a horizon, which is a 2D array, we need a 2D kernel. This is the name for the tile, or convolutional operator, we will run over the surface to change it. We will start with the simplest filter: a 3 &times; 3 mean filter. All we need is an array of 1s, which we divide by 9 for the mean.

```python
kernel = numpy.ones((3,3))
kernel
```

A visualization of the kernel is, of course, rather boring:

```python
plt.imshow(kernel, interpolation='nearest')
plt.show()
```

Smooth the data & inspect the output
===============
To  use the filter, we simply run it over the surface, one sample at a time. We simply sum the product of the values in the kernel with those in the surface. We'll have to use a nested loop to run across the columns, then down the rows, 'reading' the input with our 'viewfinder' kernel, and 'writing' the output like a book. We'll do some plus and minus 1 and 2 here and there to keep away from the edges. Hmmm, sounds complicated, maybe it's easier to do than to explain:

```python
output = numpy.empty_like(horizon)

for i in range(inlines - 2):
    for j in range(xlines - 2):
        sum_at_this_stop = 0
        for x in range(3):
            for y in range(3):
                horizon_value = horizon[(i+1)+(x-1),(j+1)+(y-1)]
                kernel_value = kernel[x,y]
                sum_at_this_stop += kernel_value * horizon_value
        output[i+1,j+1] = sum_at_this_stop / kernel.size # the mean

plt.imshow(output, cmap=plt.get_cmap('cubehelix_r'), aspect=0.5, vmin=vmin, vmax=vmax)
plt.show()
```

```python
plt.figure(figsize=(15,3))
plt.plot(horizon[300,:450])
plt.plot(noisy_horizon[300,:450], alpha=0.7)
plt.plot(output[300,:450])
plt.ylim((860,1020))
plt.legend(('original', 'noisy', 'smoothed'))
plt.gca().invert_yaxis()
plt.show()
```

It did very well, but did 'erode' the top and bottom of the fault a bit. 

Get faster
==========
If this seems a bit, er, complicated, fear not. We don't really have to go through all this. We can just use the extensive SciPy library. NumPy gives us all the *n*-dimensional array goodness, but SciPy brings the fast Fourier transforms, interpolation, linear algebra, statistics, and signal processing. So compare the crazy nested loop thing above to this:

```python
import scipy.signal
new_output = scipy.signal.convolve2d(horizon, kernel)

plt.imshow(new_output,  cmap=plt.get_cmap('cubehelix_r'), aspect=0.5, vmin=vmin, vmax=vmax)
plt.show()
```

There's yet another way to do convolution. A convolution is equivalent to a multiplication of Fourier coefficients. So we can take Fourier transforms of the kernel and the horizon, multiply them, and take the inverse Fourier transform to get our result... but this would be horrible to implement (because of various limitations of the fast Fourier transform) so let's just use the builtin

```python
fft_output = scipy.signal.fftconvolve(horizon, kernel)

plt.imshow(fft_output, cmap=plt.get_cmap('cubehelix_r'), aspect=0.5, vmin=vmin, vmax=vmax)
plt.show()
```

If you're really concerned about speed, there's an easy way to time things:

```python
%timeit new_output = scipy.signal.convolve2d(horizon, kernel)
%timeit fft_output = scipy.signal.fftconvolve(horizon, kernel)
```

Surprisingly, convolve2d is over 25% faster than fftconvolve in this case (on this data, on my machine, etc). 

More kernels
============
Convolution is a wonderful thing. I find the smoothing trick easy to understand intuitiveily — we're mixing the samples in a neighbourhood at each pixel. It's easy to see how this blurs, or smooths, the image. But look at these filters, all of a kind, and see if you can guess what they do (or maybe you recognize their names?)... 

```python
laplacian = numpy.array([[0, -1,  0],
                         [-1,  4,  -1],
                         [0, -1,  0]])

robinson = numpy.array([[-1, -1, -1],
                        [ 0,  0,  0],
                        [ 1,  1,  1]])

sobel = numpy.array([[-1, 0, 1],
                     [-2, 0, 2],
                     [-1, 0, 1]])

plt.figure(figsize=(12,3))
plt.subplot(1,3,1)
plt.imshow(laplacian, interpolation='nearest')
plt.subplot(1,3,2)
plt.imshow(robinson, interpolation='nearest')
plt.subplot(1,3,3)
plt.imshow(sobel, interpolation='nearest')
plt.show()
```

```python
shaded_relief = scipy.signal.convolve2d(horizon, laplacian)
robinson_relief = scipy.signal.convolve2d(horizon, robinson)
sobel_relief = scipy.signal.convolve2d(horizon, sobel)

plt.figure(figsize=(15,3))
plt.subplot(1,3,1)
plt.imshow(shaded_relief**0.5, cmap=plt.get_cmap('Greys'), vmax=2, aspect=0.5)
plt.subplot(1,3,2)
plt.imshow(robinson_relief**0.5, cmap=plt.get_cmap('Greys'), vmax=2, aspect=0.5)
plt.subplot(1,3,3)
plt.imshow(sobel_relief**0.5, cmap=plt.get_cmap('Greys'), vmax=2, aspect=0.5)
plt.show()
```

The <code>scipy.signal</code> library contains lots of other filters. A median filter — one of the most popular non-linear edge-preserving filters, is ready to use:

```python
median = scipy.signal.medfilt2d(horizon,11)

plt.imshow(median, cmap=plt.get_cmap('cubehelix_r'), aspect=0.5, vmin=vmin, vmax=vmax)
plt.show()
```

Likewise, a Gaussian filter is easy to implement, using a related library, <code>ndimage</code> (n-dimensional image processing). The second parameter is the standard deviation, which controls how smooth the output is.

```python
import scipy.ndimage
gaussian = scipy.ndimage.filters.gaussian_filter(horizon, 3.0)

plt.imshow(gaussian, cmap=plt.get_cmap('cubehelix_r'), aspect=0.5, vmin=vmin, vmax=vmax)
plt.show()
```

Or we can make a Gaussian kernel, if we want to look at it. I've plotted it without interpolation, so you can see the 11 &times; 11 matrix:

```python
def gaussian_kernel(size, size_y=None):
    size = int(size)
    if not size_y:
        size_y = size
    else:
        size_y = int(size_y)
    x, y = numpy.mgrid[-size:size+1, -size_y:size_y+1]
    g = numpy.exp(-(x**2/float(size)+y**2/float(size_y)))
    return g / g.sum()

# Make the Gaussian by calling the function
gaussian_kernel_array = gaussian_kernel(5)

plt.imshow(gaussian_kernel_array, interpolation='nearest', cmap=plt.get_cmap('jet'))
plt.show()
```

Final analysis
==============
All of the filters seem to deal well with the random noise. Bearing in mind that we deliberately made the Gaussian filter rather aggressive compared to the others, let's compare how they all do around that fault...

```python
plt.figure(figsize=(15,5))
plt.plot(horizon[300,220:300])
plt.plot(noisy_horizon[300,220:300])
plt.plot(output[300,220:300])
plt.plot(median[300,220:300])
plt.plot(gaussian[300,220:300])
plt.legend(('original', 'noisy', 'smoothed', 'median', 'gaussian'), loc=4)
plt.gca().invert_yaxis()
plt.show()
```

As expected, the median filter does a very good job of respecting the fault, while eliminating the random noise.
